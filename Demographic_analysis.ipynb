```python
sb.pairplot(df, hue='Make', size=2.5);
```

```python
plt.figure(1,figsize=(15,7))
n = 0
for cols in ['Age','No of Dependents','Salary','Wife Salary','Total Salary','Price']:
  n += 1
  plt.subplot(1,6,n)
  sb.set(style = 'whitegrid')
  plt.subplots_adjust(hspace=0.5,wspace=0.5)
  sb.violinplot(x= cols, y = 'Make', data=df)
  plt.ylabel("Make" if n==1 else '')
  plt.title('Violin Plot')
```

Relationship between consumers age and the vehicles they purchase

```python
plt.figure(figsize=(20,6))
sb.countplot(x="Make", data=df, hue="Age")
```

Relation between number of dependents on a consumer and the vehicles they purchase


```python
plt.figure(figsize=(20,6))
sb.countplot(x="Make", data=df, hue="No of Dependents")
```

Relation between consumers education and the vehicles they purchase

```python
plt.figure(figsize=(20,6))
sb.countplot(x="Make", data=df, hue="Education")
```
